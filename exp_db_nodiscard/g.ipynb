```python
import sys; sys.path.append('..')  # using plot.py from the parent directory
import plot
options = plot.Options(
    save=False,
    plot_nothing=True,
    plot_db=True,
    plot_pressure=True,
    print_params=False,
    db_mean_interval=2,
    plot_all_pressure=True)

exp_dir='.'
exp_bs = ['4', '8', '16', '32', '64', '128', '256', '512']
```

## YCSB Workload A

```python
files = [f'{exp_dir}/ycsb_workloada.out.xz'] + \
        [f'{exp_dir}/ycsb_workloada,at3_bs{bs}_directio.out.xz' for bs in exp_bs]

options.db_xlim = [-0.05,     60.05]
options.db_ylim = [ 0   ,  40000   ]

plot.plotFiles(files, options, allfiles=f'{exp_dir}/all-ycsb_workloada')
```
